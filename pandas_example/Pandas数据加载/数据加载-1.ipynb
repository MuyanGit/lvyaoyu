读取sqlite文件

```python
import sqlite3
```

先创建数据库连接

```python
conn = sqlite3.connect('data/data.sqlite')
conn
```

在使用pd.read_sql读取

```python
# index_col='index'就是将列名为index的那一列作为索引
weather_2017 = pd.read_sql(sql='select * from Weather_2017 limit 30',
                           con=conn,
                           index_col='index')
weather_2017
```

写入sqlite文件

```python
weather_2017.to_csv('data/weather_2017.csv')
```

变成字典

```python
weather_2017.to_dict()
```

写入json文件

```python
weather_2017.to_json('data/weather_2017.json')
```

写入html文件

```python
weather_2017.to_html('data/weather_2017.html')
```

写入数据库

```python
weather_2017.to_sql(name='Weather_2020',con=conn)
```
