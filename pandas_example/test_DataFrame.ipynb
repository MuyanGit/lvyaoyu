```python
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
df = pd.read_excel(r'C:\Users\EDZ\Desktop\汉星\excel-pdf\016.xlsx',keep_default_na=False)
df['数量'] = df['数量'].astype('int64')
df['单价'] = df['单价'].astype('int64')
df['金额'] = df['金额'].astype('int64')
df['已核销'] = df['已核销'].astype('int64')
df['未核销'] = df['未核销'].astype('int64')
df['运单号'] = [i.split('-')[-1] for i in df['运单号']]
df.index = range(2,len(df.index)+2)
df.drop(index=list(df[df['费用'] == '主单费'].index), inplace=True)
del_list = []
for i in range(0,len(df.index)-1):
    for j in range(i+1,len(df.index)):
        if df.loc[df.index[i],'订单号']==df.loc[df.index[j],'订单号'] and df.loc[df.index[i],'运单号']==df.loc[df.index[j],'运单号'] and df.loc[df.index[i],'航程']==df.loc[df.index[j],'航程'] and df.loc[df.index[i],'数量']==df.loc[df.index[j],'数量'] and df.loc[df.index[i],'费用']==df.loc[df.index[j],'费用']:
            print(df.index[i],df.index[j])
            df.loc[df.index[i],'单价'] += df.loc[df.index[j],'单价']
            df.loc[df.index[i],'金额'] += df.loc[df.index[j],'金额']
            df.loc[df.index[i],'已核销'] += df.loc[df.index[j],'已核销']
            df.loc[df.index[i],'未核销'] += df.loc[df.index[j],'未核销']
            df.loc[df.index[i],'附加说明'] = df.loc[df.index[i],'附加说明'] if df.loc[df.index[i],'附加说明']!='' else df.loc[df.index[j],'附加说明']
            del_list.append(df.index[j])

df.drop(index=del_list,inplace=True)
df.to_excel(r'C:\Users\EDZ\Desktop\汉星\excel-pdf\副本-016.xlsx')
```
